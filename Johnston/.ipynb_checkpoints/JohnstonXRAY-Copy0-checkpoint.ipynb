## Energy Calibration

```python
elements = ['Fe K-a', 'Fe K-b', 'Cu K-a', 'Cu K-b', 'Ni K-a', 'Ni K-b', 'Zn K-a', 'Zn K-b', \
            'Zr K-a', 'Zr K-b', 'Mo K-a', 'Mo K-b', 'Rh K-a', 'Rh K-b', 'Sn K-a', 'Sn K-b', \
            'Pb L-a', 'Pb L-b' ]
channels = np.array([66.33, 73.07, 83.32, 92.22, 77.43, 85.61, 89.36, 99.05, \
                  162.94, 182.80, 180.55, 202.90, 208.59, 235.00, 260.56, 294.27, \
                  109.20, 130.69])
FWHM = np.array([2.296, 2.284, 2.319, 2.311, 2.392, 2.366, 2.473, 2.449, \
              2.903, 2.618, 3.093, 2.764, 3.307, 2.973, 4.004, 2.979, \
              2.635, 2.626])
energy = np.array([6.40, 7.06, 8.05, 8.90, 7.48, 8.26, 8.64, 9.57, \
                15.77, 17.67, 17.48, 19.61, 20.21, 22.72, 25.27, 28.48, \
                10.55, 12.61])

#Import the curve fitter from the scipy optimize package
from scipy.optimize import curve_fit

#Define the fit function
func = lambda x,m,b: m*x + b

# Make initial guess at parameters
p0 = [0.1, -0.05]

#Call the curve fitter with FWHM as standard weights
#and have it return the optimized parameters (popt) and covariance matrix (pcov)
popt,pcov = curve_fit(func,channels,energy,p0,sigma=1/FWHM**2)

fitresult = func(channels,popt[0],popt[1])
print popt[0],popt[1]
```

```python
plt.plot(channels,energy,"ro",label="Calibration data")
curve="Energy = %.3f*Channel + %.3f"%(popt[0],popt[1])
plt.plot(channels,fitresult,'b-',label=curve)
plt.xlabel("Channel Number",fontsize=20)
plt.ylabel("Energy (keV)",fontsize=20)
plt.legend(loc="best")
plt.plot()
```

## Apply calibration

```python
#convert FWHM from channels to energy with calibration equation
calFWHM = FWHM*popt[0] + popt[1]
#convert channels to energy with calibration equation
calEn = channels*popt[0] + popt[1]
```

## Energy resolution

```python
func2 = lambda x,a,b,c: a + b*x + c*x**-2
popt2,pcov2 = curve_fit(func2,calEn,calFWHM/calEn,[0.02,-1.e-3,0.5])
print popt2
x = np.linspace(0.,40.)
resolution = func2(x,popt2[0],popt2[1],popt2[2])
print resolution[3.685]*3.685
print resolution[4.014]*4.014

plt.plot(calEn,calFWHM/calEn,"ro",label="Data")
plt.plot(x,resolution,'b-',label="a + b*x + c*x**-2")
plt.xlabel("Energy (keV)",fontsize=20)
plt.ylabel("FWHM/Energy (calibrated)",fontsize=20)
plt.legend(loc="best")
plt.ylim(0.,0.05)
plt.plot()

```

### Populate a list of lists called difteq that holds the energies of two peaks in a tuple, the best fit element for the peaks, and an expression for the error of the guess

```python
difference = [[],[],[]]#energy1, energy2, difference
difteq = [[],[],[],[],[],[]]#(energy1K, energy2K),(energy1L, energy2L),elementK,elementL, K error value, L error value
newel = False
kgot = False
d = .5
pos = .2
def peakfind(d,shift):
    difference[:]=[[],[],[]]
    difteq[:]=[[],[],[],[],[],[]]
    for i in xrange(len(peak[0])):
        for p in xrange(i,len(peak[0])-1):
            difference[0].append(peak[0][i]+shift)
            difference[1].append(peak[0][p+1]+shift)
            difference[2].append(peak[0][p+1]-peak[0][i])
    for j in xrange(len(difference[2])):
        distance = 1000
        newel = True
        kgot = False
        for ka in KLines.keys():
            if abs(difference[2][j]-(KLines[ka][1]-KLines[ka][0])) < distance and abs(KLines[ka][0]-difference[0][j]) <= d and \
            abs(KLines[ka][1]-difference[1][j]) <= d:
                #print ka,j,difference[2][j]-(K[ka][1]-K[ka][0]), "k"
                if newel == False:
                    del difteq[2][-1]
                    del difteq[0][-1]
                    del difteq[4][-1]
                distance = abs(difference[2][j]-(KLines[ka][1]-KLines[ka][0]))
                difteq[2].append(ka)
                difteq[4].append(distance+abs(KLines[ka][0]-difference[0][j])+abs(KLines[ka][1]-difference[1][j]))
                difteq[0].append((difference[0][j],difference[1][j]))
                kgot = True
                newel = False
                #print difteq[2]
        distance = 1000
        newel = True
        for la in LLines.keys():
            if abs(difference[2][j]-(LLines[la][1]-LLines[la][0]))<distance and abs(LLines[la][0]-difference[0][j]) <= d and \
            abs(LLines[la][1]-difference[1][j]) <= d:
                #print la, j,difference[2][j]-(L[la][1]-L[la][0]), "l"
                if newel == False:
                    del difteq[3][-1]
                    del difteq[1][-1]
                    del difteq[5][-1]
                distance = abs(difference[2][j]-(LLines[la][1]-LLines[la][0]))
                if kgot == True and difteq[4][-1]>distance+abs(LLines[la][0]-difference[0][j])+abs(LLines[la][1]-difference[1][j]):
                    difteq[3].append(la)
                    difteq[5].append(distance+abs(LLines[la][0]-difference[0][j])+abs(LLines[la][1]-difference[1][j]))
                    difteq[1].append((difference[0][j],difference[1][j]))
                    del difteq[2][-1]
                    del difteq[0][-1]
                    del difteq[4][-1]
                    #print difteq[2]
                    newel = False
                if kgot == False:
                    difteq[3].append(la)
                    difteq[5].append(distance+abs(LLines[la][0]-difference[0][j])+abs(LLines[la][1]-difference[1][j]))
                    difteq[1].append((difference[0][j],difference[1][j]))
                    newel = False
                #print difteq[3]
    #print difteq
    #print len(difteq[2]), len(difteq[3]), len(difteq[0])
    i=0
    while i+1 < len(difteq[0]):
        if difteq[4][i+1]<difteq[4][i] and difteq[0][i+1][0]==difteq[0][i][0]:
            del difteq[2][i]
            del difteq[4][i]
            del difteq[0][i]
        elif difteq[4][i+1]>difteq[4][i] and difteq[0][i+1][0]==difteq[0][i][0]:
            del difteq[2][i+1]
            del difteq[4][i+1]
            del difteq[0][i+1]
        else:
            i += 1
    #print difteq
    j=0
    while j+1 < len(difteq[1]):
        if difteq[5][j+1]<difteq[5][j] and difteq[1][j+1][0]==difteq[1][j][0]:
            del difteq[3][j]
            del difteq[5][j]
            del difteq[1][j]
        elif difteq[5][j+1]>difteq[5][j] and difteq[1][j+1][0]==difteq[1][j][0]:
            del difteq[3][j+1]
            del difteq[5][j+1]
            del difteq[1][j+1]
        else:
            j += 1
    #print difteq
```

### This cell is my crowning glory. It optimizes the error value given when we guess our elements for the pairs of peaks. It then varies the shift parameter from the previous cell to find the lowest average of all the error values in our set of guessed elements.

```python
'''Optimizing error values by varying shift variable until i get the best average error value.'''
def errave(): #In this function I average the erors of each shift. I am trying to add weights based on height of peak now
    errave = 0
    erravek = 0
    erravel = 0
    #wsumk = 0
    #wsuml = 0
    #for k in xrange(len(difteq[4])):
        #wsumk += ((MultCts[np.searchsorted(calEn,difteq[0][k][0])]+MultCts[np.searchsorted(calEn,difteq[0][k][1])])/2)
        #print wsumk
    #for w in xrange(len(difteq[4])):
       # wsumk += ((MultCts[np.searchsorted(calEn,difteq[0][w][0])]+MultCts[np.searchsorted(calEn,difteq[0][w][1])])/2)
    for i in xrange(len(difteq[4])):
        erravek += difteq[4][i]#*((MultCts[np.searchsorted(calEn,difteq[0][i][0])]+MultCts[np.searchsorted(calEn,difteq[0][i][1])])/2)
    erravek = erravek/len(difteq[4])
    for l in xrange(len(difteq[5])):
        erravel += difteq[5][l]#*((MultCts[np.searchsorted(calEn,difteq[1][i][0])]+MultCts[np.searchsorted(calEn,difteq[1][i][1])])/2)
    erravel = erravel/len(difteq[5])
    #erravek = erravek/wsumk
    #erravel = erravel/wsuml
    errave = (erravek+erravel)/2
    return errave
    
d = .5
erraverage = 0
besterrave = 10000
j=-2
bestshift = 0
while j<=2:
    try:
        peakfind(d, j)
        erraverage = errave()
        if erraverage<besterrave:
            besterrave = erraverage
            bestshift = j
            print j,besterrave
    except IndexError:
            j+=.001
    j+=.001
    
    
```

### This cell defines a function to trim the Counts/s if they are below some user defined threshhold

```python
def trim(threshhold):
    for i in xrange(len(MultCts)):
        if MultCts[i] < threshhold:
            MultCts[i] = 0
```

### This cell plots our data with variable sliders to zoom in, set our theshold, and execute the trim by setting the trim slider to one

```python
def identify(threshhold = .02, zoom = 1, trimit = 0):
    fig = figure(2,(24,16),dpi=72)
    if trimit == 1:
        trim(threshhold)
    fig.clear()
    plt.plot(calEn, MultCts,'g-')
    plt.plot((0,40), (threshhold,threshhold), 'b--')
    plt.xlim(0.,40.)
    plt.legend(fontsize=20)
    plt.xlabel("Energy (keV)",fontsize=20)
    plt.ylabel("Counts/s",fontsize=20)
    plt.ylim(0,zoom)
    plt.show()
    
v = interact(identify, threshhold=(0.00,.050,.0001), zoom = (0,2,.001), trimit =(0,1))
```

```python
def iron(x,a0,a1,sig): 
    """Return the integrated area and common
    width of Gaussians for the K-alpha and 
    K-beta lines of iron"""
    
    #use the resolution curve to set the width?
    ampKa = a0/(sig*np.sqrt(2*np.pi))
    ampKb = a1/(sig*np.sqrt(2*np.pi))
    menKa = KLines['Fe'][0]
    menKb = KLines['Fe'][1]
    #print menKa,menKb
    return ampKa*np.exp(-(x-menKa)**2/(2*sig**2))+ampKb*np.exp(-(x-menKb)**2/(2*sig**2))

popt, pcov = curve_fit(iron, calEn[58:80], ct1[58:80], p0 = [10.,15.,0.1])
#print popt

fig = figure(2,(12,8),dpi=72)
plt.plot(calEn[0:380],ct1[0:380],'g-')
plt.xlim(0.,40.)
plt.ylim(0.,)
plt.plot(calEn[0:380],iron(calEn[0:380], *popt))
#plt.plot(calEn[0:380],iron(calEn[0:380],10.,15.,0.1),"r-")
plt.show()
```
