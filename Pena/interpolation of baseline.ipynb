```python
data = np.loadtxt("../data/Pena-specimen-12Feb2014.txt")
norm = 1994.569 #amount of time the experiment was run

chan = np.arange(512)
calEn = chan*popt[0] + popt[1]#
#print calEn
cts = data/norm
```

```python
'''
Here I find all local mins and define boundaries as horizontal lines of connecting to closest value.
I then interpolate the data so that it is defined at any point that we wish. This is so that it is defined at all energy values.
'''
low = [[],[]]
init = False
for en in xrange(2,511):
    if cts[en]<cts[en+1] and cts[en]<cts[en-1]:
        if init == False:
            low[0].append(calEn[0])
            low[1].append(cts[en])
            init = True
        low[0].append(calEn[en])
        low[1].append(cts[en])
low[0].append(calEn[-1])
low[1].append(low[1][-1])
print low
lowf = interp1d(low[0],low[1], kind="linear")
print lowf(0)
print calEn
```

```python
'''
Here I have plotted our spectrum, the raw baseline data(from previous cell) and the interpolation of the baseline.
'''
fig = figure(1,(12,8),dpi=72)
plt.plot(calEn,cts,'r-',linewidth=2,label="Pena Specimen")
plt.plot(low[0], low[1],'b-',linewidth=2,label="Baseline")
plt.plot(calEn, lowf(calEn),'g-',linewidth=2,label="Baseline Interpolated")
plt.plot(calEn,cts**3,'y-',linewidth=2,label="Pena Specimen^2")
plt.xlabel("Energy (keV)",fontsize=20)
plt.ylabel("Counts/s",fontsize=20)
plt.xlim(12,20.)
plt.ylim(0.,2.)
plt.legend(fontsize=20)
plt.show()
```

```python
'''
This cell plots the difference between our data and the interpolated baseline
'''
fig = figure(2,(12,8),dpi=72)
plt.plot(calEn, cts-lowf(calEn),'y-',linewidth=2,label="Pena-Baseline")
plt.xlabel("Energy (keV)",fontsize=20)
plt.ylabel("Counts/s",fontsize=20)
plt.xlim(0.,30.)
plt.ylim(-.05, .05) #I used this line as a tool to decide threshhold
plt.legend(fontsize=20)
plt.show()
```
